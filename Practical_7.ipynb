Aim:- Implement and analyze various probability distributions in Python, including Binomial, Poisson, and Normal distributions.

### 1. What is the difference between Binomial, Poisson, and Normal distributions?

*   **Binomial Distribution:** This is a **discrete** probability distribution that models the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes (success or failure) and the probability of success is constant for each trial.
    *   **Key Characteristics:** Fixed number of trials, independent trials, two outcomes, constant probability of success.
    *   **Example:** The number of heads in 10 coin flips.

*   **Poisson Distribution:** This is also a **discrete** probability distribution that models the number of events occurring within a fixed interval of time or space, given a constant average rate of occurrence and that these events happen independently.
    *   **Key Characteristics:** Events occur at a constant average rate, events are independent, interval is fixed (time, space, etc.).
    *   **Example:** The number of calls received by a call center in an hour.

*   **Normal Distribution (Gaussian Distribution):** This is a **continuous** probability distribution that is symmetric around its mean, forming a bell-shaped curve. It's characterized by its mean (average) and standard deviation (spread).
    *   **Key Characteristics:** Continuous data, symmetric, bell-shaped, described by mean and standard deviation.
    *   **Example:** Human height, blood pressure, measurement errors.

### 2. What is the real-world significance of the Poisson distribution?

The Poisson distribution is significant for modeling the number of occurrences of rare events over a specified interval when the average rate of occurrence is known and events happen independently. Its real-world applications include:

*   **Healthcare:** Modeling the number of patients arriving at an emergency room per hour.
*   **Manufacturing:** Counting the number of defects in a roll of fabric or a batch of products.
*   **Traffic Management:** Predicting the number of cars passing a certain point on a road per minute.
*   **Insurance:** Estimating the number of claims an insurance company might receive in a given period.
*   **Ecology:** Analyzing the number of mutations in a DNA strand or the number of rare species observed in a region.

### 3. What is the Central Limit Theorem and how does it relate to Normal distribution?

*   **Central Limit Theorem (CLT):** The CLT states that, regardless of the original distribution of a population, if you take sufficiently large random samples from that population, the distribution of the sample means will be approximately normally distributed. This approximation improves as the sample size increases.

*   **Relation to Normal Distribution:** The CLT is fundamental because it explains why the Normal distribution is so prevalent in statistics. Even if the data itself is not normally distributed (e.g., skewed, uniform), the sampling distribution of the mean will tend towards normal. This allows statisticians to use powerful statistical methods (like hypothesis testing and confidence intervals) that rely on the properties of the Normal distribution, even when dealing with non-normal raw data, provided the sample size is large enough.

### 4. Why is Normal distribution widely used in machine learning and statistics?

The Normal distribution is widely used due to several key reasons:

*   **Central Limit Theorem:** As explained above, it's the distribution that sample means tend towards, making it crucial for inferential statistics.
*   **Mathematical Tractability:** It's well-understood mathematically, simplifying calculations and allowing for analytical solutions in many statistical models.
*   **Parameter Estimation:** Its parameters (mean and standard deviation) are often easy to estimate from data, and these estimates are often optimal under certain conditions.
*   **Ubiquity in Nature and Data:** Many natural phenomena and measurement errors naturally follow an approximately normal distribution.
*   **Foundation for Many Algorithms:** Many statistical and machine learning algorithms (e.g., Linear Regression, ANOVA, Gaussian Naive Bayes, many hypothesis tests) assume or perform optimally when data, or errors, are normally distributed.
*   **Interpretation:** Its symmetry and bell shape make it intuitive for describing the typical value and variability of data.

### 5. What is skewness in distributions and which distributions exhibit it?

**Skewness** is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. It indicates the direction and magnitude of the departure from symmetry.

*   **Types of Skewness:**
    *   **Positive Skew (Right-Skewed):** The tail of the distribution is longer on the right side. The mean is typically greater than the median.
    *   **Negative Skew (Left-Skewed):** The tail of the distribution is longer on the left side. The mean is typically less than the median.
    *   **Zero Skew (Symmetric):** The distribution is perfectly symmetrical, and the mean, median, and mode are often equal (e.g., Normal distribution).

**Distributions that exhibit skewness:**

*   **Exponential Distribution:** Always positively skewed.
*   **Log-Normal Distribution:** Always positively skewed.
*   **Chi-squared Distribution:** Positively skewed, especially with fewer degrees of freedom.
*   **Binomial Distribution:** Can be skewed if the probability of success `p` is not 0.5 (positively skewed if `p < 0.5`, negatively skewed if `p > 0.5`).
*   **Poisson Distribution:** Positively skewed, particularly for smaller mean values (λ). As λ increases, it becomes more symmetric and approaches a Normal distribution.
*   **Weibull Distribution:** Can be either positively or negatively skewed depending on its shape parameter.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm

# Set up a figure with subplots
fig, axs = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle("Analysis of Binomial, Poisson, and Normal Distributions", fontsize=16)

# ------------------- BINOMIAL DISTRIBUTION -------------------
n, p = 10, 0.5
x_binom = np.arange(0, n+1)
pmf_binom = binom.pmf(x_binom, n, p)
cdf_binom = binom.cdf(x_binom, n, p)
samples_binom = binom.rvs(n, p, size=1000)

# PMF
axs[0, 0].bar(x_binom, pmf_binom, color='skyblue')
axs[0, 0].set_title("Binomial PMF (n=10, p=0.5)")
# CDF
axs[0, 1].step(x_binom, cdf_binom, where='mid', color='blue')
axs[0, 1].set_title("Binomial CDF")
# Histogram of Samples
axs[0, 2].hist(samples_binom, bins=np.arange(n+2)-0.5, density=True, color='navy', alpha=0.7)
axs[0, 2].set_title("Binomial Samples Histogram")

# ------------------- POISSON DISTRIBUTION -------------------
mu = 4
x_poisson = np.arange(0, 15)
pmf_poisson = poisson.pmf(x_poisson, mu)
cdf_poisson = poisson.cdf(x_poisson, mu)
samples_poisson = poisson.rvs(mu, size=1000)

# PMF
axs[1, 0].bar(x_poisson, pmf_poisson, color='lightgreen')
axs[1, 0].set_title("Poisson PMF (μ=4)")
# CDF
axs[1, 1].step(x_poisson, cdf_poisson, where='mid', color='green')
axs[1, 1].set_title("Poisson CDF")
# Histogram of Samples
axs[1, 2].hist(samples_poisson, bins=np.arange(16)-0.5, density=True, color='darkgreen', alpha=0.7)
axs[1, 2].set_title("Poisson Samples Histogram")

# ------------------- NORMAL DISTRIBUTION -------------------
mu_norm, sigma = 0, 1
x_norm = np.linspace(-4, 4, 1000)
pdf_norm = norm.pdf(x_norm, mu_norm, sigma)
cdf_norm = norm.cdf(x_norm, mu_norm, sigma)
samples_norm = norm.rvs(mu_norm, sigma, size=1000)

# PDF
axs[2, 0].plot(x_norm, pdf_norm, color='salmon')
axs[2, 0].set_title("Normal PDF (μ=0, σ=1)")
# CDF
axs[2, 1].plot(x_norm, cdf_norm, color='red')
axs[2, 1].set_title("Normal CDF")
# Histogram of Samples
axs[2, 2].hist(samples_norm, bins=30, density=True, color='darkred', alpha=0.7)
axs[2, 2].set_title("Normal Samples Histogram")

# Layout and show
for row in axs:
    for ax in row:
        ax.grid(True)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# ------------------- BASIC STATISTICAL ANALYSIS -------------------
print("Binomial Distribution:")
print(f"Mean = {np.mean(samples_binom):.2f}, Variance = {np.var(samples_binom):.2f}")
print("\nPoisson Distribution:")
print(f"Mean = {np.mean(samples_poisson):.2f}, Variance = {np.var(samples_poisson):.2f}")
print("\nNormal Distribution:")
print(f"Mean = {np.mean(samples_norm):.2f}, Standard Deviation = {np.std(samples_norm):.2f}")
```
