```python
import pandas as pd
import numpy as np

# 1. Create a sample DataFrame
data = {
    'ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi'],
    'Age': [24, 27, 22, 32, np.nan, 29, 31, 25],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'London', 'New York', 'Paris', 'Berlin'],
    'Salary': [50000, 60000, 45000, 75000, 55000, np.nan, 62000, 48000]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
display(df)

# 2. Inspect data (understanding data types and missing values)
print("\nDataFrame Info:")
df.info()

# 3. Handle missing values
# Fill missing 'Age' values with the mean age
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Drop rows with any missing values (e.g., 'Salary')
df.dropna(inplace=True)

print("\nDataFrame after handling missing values:")
display(df)

# 4. Data type conversion (if necessary, though in this case, pandas inferred correctly)
# Example: Convert 'Age' to integer after filling NaN, if it became float
df['Age'] = df['Age'].astype(int)

# 5. Transform data: Create a new 'Salary_USD_Thousands' column
df['Salary_USD_Thousands'] = df['Salary'] / 1000

# 6. Transform data: Categorical encoding (one-hot encoding for 'City')
df_encoded = pd.get_dummies(df, columns=['City'], drop_first=True)

print("\nDataFrame after transformations (new column and one-hot encoding):")
display(df_encoded)

# 7. Basic filtering
print("\nDataFrame with people older than 25:")
display(df[df['Age'] > 25])
```

### 1. What is data wrangling? Explain the difference between data wrangling and data preprocessing.

**Data Wrangling** (also known as data munging) is the process of cleaning, structuring, and enriching raw data into a desired format for better decision-making and analysis. It involves various tasks like handling missing values, standardizing formats, correcting inconsistencies, and integrating data from multiple sources. It often requires iterative exploration and transformation.

**Data Preprocessing** is a broader term that encompasses all steps taken to prepare raw data for a machine learning model or analysis. It includes:
*   **Cleaning:** Handling missing values, noise, outliers.
*   **Transformation:** Normalization, standardization, feature scaling, encoding categorical data.
*   **Reduction:** Dimensionality reduction (e.g., PCA), numerosity reduction.
*   **Discretization:** Converting continuous attributes into discrete ones.

**Key Difference:**
*   **Scope:** Data preprocessing is a more general term for preparing data for analytical or machine learning tasks. Data wrangling is a specific, often iterative and exploratory, part of data preprocessing that focuses on shaping and structuring raw data into a usable form, often involving more complex transformations and integration.
*   **Objective:** Wrangling aims to make data usable and understandable for initial analysis or further processing. Preprocessing aims to optimize data for specific models or analyses, which might include steps like scaling that aren't strictly part of 'wrangling'.

### 2. What is the importance of exploratory data analysis (EDA) in Pandas?

**Exploratory Data Analysis (EDA)** using Pandas is crucial for several reasons:
*   **Understanding Data:** It helps in gaining insights into the dataset's structure, content, and statistical properties (e.g., central tendency, spread, distributions).
*   **Identifying Patterns and Relationships:** EDA helps uncover trends, patterns, correlations, and relationships between variables that might not be obvious at first glance.
*   **Detecting Anomalies and Outliers:** Visualizations and statistical summaries can highlight unusual data points or errors that need to be addressed.
*   **Formulating Hypotheses:** Insights from EDA can lead to the formulation of hypotheses for further investigation or modeling.
*   **Feature Engineering:** Understanding data distributions and relationships can guide decisions on creating new features or transforming existing ones.
*   **Data Cleaning Guidance:** EDA helps identify missing values, inconsistent formats, and data types that need correction, guiding the data cleaning process.
*   **Validating Assumptions:** It allows checking assumptions about the data that might be critical for certain statistical tests or machine learning models.

Pandas provides excellent tools (e.g., `info()`, `describe()`, `value_counts()`, `groupby()`, various plotting functions) that make EDA efficient and effective.

### 3. What are the main differences between Series, DataFrame, and Panel?

Pandas offers three primary data structures:

*   **Series:**
    *   **Dimensionality:** 1-dimensional labeled array.
    *   **Homogeneity:** Can hold data of any single data type (integer, float, string, Python objects, etc.).
    *   **Analogy:** Similar to a column in a spreadsheet or a SQL table, or a NumPy array with an index.
    *   **Components:** Data + Index.
    *   **Use Case:** Representing a single feature or variable.

*   **DataFrame:**
    *   **Dimensionality:** 2-dimensional labeled data structure.
    *   **Homogeneity:** Can hold columns of potentially different data types.
    *   **Analogy:** Similar to a spreadsheet, a SQL table, or a dictionary of Series objects.
    *   **Components:** Data (rows and columns) + Row Index + Column Index.
    *   **Use Case:** Representing a tabular dataset with multiple features.

*   **Panel (Deprecated):**
    *   **Dimensionality:** 3-dimensional labeled data structure.
    *   **Homogeneity:** Could hold items of potentially different data types.
    *   **Analogy:** Can be thought of as a dictionary of DataFrames, where each DataFrame corresponds to an 'item'.
    *   **Components:** Items + Major Axis (rows) + Minor Axis (columns).
    *   **Use Case (originally):** Handling three-dimensional data, e.g., time series data across multiple entities. However, due to its complexity and the availability of better alternatives (MultiIndex DataFrames, xarray), `Panel` has been deprecated since Pandas version 0.20.0 and removed in 1.0.0. For 3D data, it's recommended to use MultiIndex DataFrames or `xarray` instead.

### 4. Explain the use of:

a.  **`isnull()`**
    *   **Purpose:** This method returns a boolean same-sized object (Series or DataFrame) indicating whether each element is `NaN` (Not a Number) or `None` (missing value).
    *   **Output:** `True` for missing values, `False` for non-missing values.
    *   **Example:** `df.isnull()` or `df['Column'].isnull()`

b.  **`notnull()`**
    *   **Purpose:** The inverse of `isnull()`. It returns a boolean same-sized object indicating whether each element is *not* `NaN` or `None`.
    *   **Output:** `True` for non-missing values, `False` for missing values.
    *   **Example:** `df.notnull()` or `df['Column'].notnull()`

c.  **`dropna()`**
    *   **Purpose:** This method is used to remove rows or columns that contain missing values (`NaN` or `None`).
    *   **Parameters:**
        *   `axis`: `0` (default) for rows, `1` for columns.
        *   `how`: `'any'` (default) to drop if *any* `NaN` is present, `'all'` to drop only if *all* values are `NaN`.
        *   `inplace`: `False` (default) to return a new DataFrame, `True` to modify the DataFrame in place.
    *   **Example:** `df.dropna()`, `df.dropna(axis=1)`, `df.dropna(how='all', inplace=True)`

d.  **`fillna()`**
    *   **Purpose:** This method is used to replace missing values (`NaN` or `None`) with a specified value or method.
    *   **Parameters:**
        *   `value`: A scalar value (e.g., 0, 'missing'), a dictionary, or a Series to fill missing values.
        *   `method`: `'ffill'` (forward fill, propagates last valid observation forward) or `'bfill'` (backward fill, propagates next valid observation backward).
        *   `axis`: `0` (default) for filling along rows, `1` for filling along columns.
        *   `inplace`: `False` (default) to return a new DataFrame, `True` to modify the DataFrame in place.
    *   **Example:** `df['Age'].fillna(df['Age'].mean())`, `df.fillna(0)`, `df.fillna(method='ffill')`

### 5. What is column transformation and why is it important in data analysis?

**Column transformation** in data analysis refers to the process of modifying the values within one or more columns of a dataset to create new columns, change their data types, or normalize their distributions. These transformations can be simple (e.g., changing units) or complex (e.g., creating polynomial features).

**Importance in Data Analysis:**
*   **Feature Engineering:** Creating new features from existing ones can significantly improve the performance of machine learning models. For example, extracting `day_of_week` from a `date` column.
*   **Handling Skewness and Outliers:** Transformations like logarithmic or square root transformations can make skewed data more normally distributed, which is beneficial for many statistical models and algorithms that assume normality.
*   **Normalization and Scaling:** Bringing features to a similar scale prevents features with larger values from dominating the learning process. This is crucial for algorithms sensitive to feature magnitudes (e.g., K-Nearest Neighbors, Support Vector Machines).
*   **Categorical Encoding:** Converting categorical variables (e.g., 'City', 'Gender') into numerical representations (e.g., one-hot encoding, label encoding) is necessary for most machine learning algorithms that require numerical input.
*   **Data Type Conversion:** Ensuring columns have the correct data type (e.g., converting a string '25' to an integer 25) is fundamental for accurate calculations and memory efficiency.
*   **Data Cleaning and Consistency:** Transforming data can help in standardizing formats, correcting inconsistencies, or creating flags for specific conditions.
*   **Simplifying Relationships:** Sometimes, transforming variables can linearize non-linear relationships, making them easier to model and interpret.

In essence, column transformations are a critical step in preparing data for analysis and modeling, allowing analysts to extract more information, improve model performance, and ensure data compatibility with various analytical techniques.
