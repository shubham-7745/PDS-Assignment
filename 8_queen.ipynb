Assignment on 8 queen

```python
N = 8 # size of the chessboard

# List to store all solutions
solutions = []

# Function to print the board
def print_board(board):
    for row in board:
        print(" ".join(row))
    print()

# Check if placing a queen at board[row][col] is safe
def is_safe(board, row, col):
    # Check column
    for i in range(row):
        if board[i][col] == "Q":
            return False

    # Check upper left diagonal
    i, j = row - 1, col - 1
    while i >= 0 and j >= 0:
        if board[i][j] == "Q":
            return False
        i -= 1
        j -= 1

    # Check upper right diagonal
    i, j = row - 1, col + 1
    while i >= 0 and j < N:
        if board[i][j] == "Q":
            return False
        i -= 1
        j += 1
    return True

# Solve function using recursion
def solve(board, row):
    if row == N: # all queens placed
        # Found a solution, add a deep copy to the solutions list
        solutions.append([row[:] for row in board])
        return

    for col in range(N):
        if is_safe(board, row, col):
            board[row][col] = "Q" # place queen
            solve(board, row + 1) # place next queen
            board[row][col] = "." # backtrack

# Main code
board = [["." for _ in range(N)] for _ in range(N)]
solve(board, 0)

# Print all found solutions
if solutions:
    print(f"Found {len(solutions)} solutions for {N}-Queens:\n")
    for i, sol_board in enumerate(solutions):
        print(f"Solution {i + 1}:")
        print_board(sol_board)
else:
    print(f"No solutions found for {N}-Queens.")
```
