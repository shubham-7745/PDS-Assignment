## Theory

### 1. What is web scraping? Explain the role of the Requests library in web scraping.
*   **Web scraping** is the process of automatically extracting information or data from websites. It involves programmatically fetching web pages and parsing their content to retrieve specific data points. This data can then be stored, analyzed, or used for various purposes like market research, price comparison, data analytics, and more.
*   The **Requests library** in Python is crucial for web scraping because it handles the initial step of making HTTP requests to a website. It allows you to send various types of HTTP requests (GET, POST, PUT, DELETE, HEAD, OPTIONS) to a specified URL and retrieve the server's response. In web scraping, `requests.get(url)` is commonly used to fetch the HTML content of a web page, which is the raw data that will then be parsed.

### 2. What is BeautifulSoup? How does it help in parsing HTML?
*   **Beautiful Soup** is a Python library designed for parsing HTML and XML documents. It creates a parse tree from the page source code, which can be navigated, searched, and modified. It's often used in conjunction with the Requests library.
*   **How it helps in parsing HTML:** After Requests fetches the raw HTML content of a web page, BeautifulSoup takes this raw HTML as input. It then turns the complex and unstructured HTML into a Python object (a parse tree) that represents the page's structure. This allows you to:
    *   **Navigate the parse tree:** Easily move through the HTML structure using tag names (e.g., `soup.body`, `soup.p`).
    *   **Search for specific elements:** Find elements based on their tag name, attributes (like `id` or `class`), or text content using methods like `find()` and `find_all()`.
    *   **Extract data:** Access the text content, attribute values, or nested tags of the found elements.
    *   **Handle malformed HTML:** Beautiful Soup is robust and can handle poorly structured or broken HTML, making it very practical for real-world web pages.

### 3. Differentiate between web scraping and web crawling.
*   **Web Scraping:** Focuses on extracting specific data from specific web pages. You usually know what data you want and from which pages. It's like going to a specific page in a library and copying down a particular piece of information.
*   **Web Crawling:** Focuses on discovering and indexing web pages by following hyperlinks. A web crawler (or spider) starts with a seed URL, retrieves the page, extracts all the links on that page, and then visits those links recursively. Its goal is to explore as much of the web as possible, often to build search engine indexes. It's like wandering through a library, looking at all the book titles, and noting down which books are linked to others.
*   **Relationship:** Web scraping often utilizes web crawling to find the pages from which to scrape data. A crawler finds the URLs, and a scraper extracts data from those URLs.

### 4. What are the commonly used methods in BeautifulSoup?
Some of the most commonly used methods in BeautifulSoup include:
*   `BeautifulSoup(html_doc, 'html.parser')`: The constructor to create a `BeautifulSoup` object from HTML content.
*   `.find(name, attrs, string, **kwargs)`: Finds the first tag that matches the given criteria.
*   `.find_all(name, attrs, string, limit, **kwargs)`: Finds all tags that match the given criteria and returns them as a list. `limit` can restrict the number of results.
*   `.select(selector)`: Allows you to search for elements using CSS selectors, which can be very powerful (e.g., `soup.select('div.product-name a')`).
*   `.get_text()`: Extracts all the text content within a tag, stripping out HTML tags.
*   `['attribute_name']` or `.get('attribute_name')`: Accesses the value of an attribute of a tag (e.g., `link['href']`).
*   `.parent`, `.children`, `.next_sibling`, `.previous_sibling`: Methods for navigating the parse tree relative to a tag.

### 5. What are the ethical and legal considerations in web scraping?
Web scraping, while powerful, comes with significant ethical and legal considerations:
*   **Terms of Service (ToS):** Many websites explicitly prohibit scraping in their terms of service. Violating ToS can lead to legal action, account termination, or IP blocking.
*   **Copyright and Intellectual Property:** The data you scrape might be copyrighted. Copying and republishing large portions of data without permission can infringe on intellectual property rights.
*   **Data Privacy:** Scraping personal identifiable information (PII) is a major concern. Regulations like GDPR, CCPA, and others strictly govern how personal data can be collected, processed, and stored. Scraping PII without consent is generally illegal.
*   **Server Load and Denial of Service:** Aggressive scraping can put a heavy load on a website's server, potentially slowing it down or even causing a denial of service (DoS). This is unethical and can be illegal.
*   **`robots.txt`:** Websites often use a `robots.txt` file to specify which parts of their site should not be crawled or scraped by bots. While not legally binding in all jurisdictions, ignoring `robots.txt` is generally considered unethical and can be used as evidence against you in a legal dispute.
*   **Deep Linking:** Some websites prohibit deep linking (linking directly to internal pages rather than the homepage). While scraping doesn't always involve linking, it relates to how you access and use the site's structure.
*   **Transparency and Attribution:** Ethically, it's good practice to be transparent about your scraping activities (if possible and safe to do so) and to attribute the source of your data if you publish or use it.

**Best Practices (Ethical & Legal):**
*   Always check the website's `robots.txt` file.
*   Review the website's Terms of Service.
*   Be respectful of server load (e.g., use delays between requests).
*   Do not scrape personal or sensitive data.
*   Consider if the data is publicly available through an API first. If so, use the API.
*   Seek legal advice if you are unsure about the legality of your scraping activities, especially for commercial use or large-scale projects.

```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the website to scrape
# Using a sample public domain website for demonstration purposes
# This website is specifically designed for web scraping practice.
url = 'http://quotes.toscrape.com/'

print(f"Attempting to fetch data from: {url}")

# Send an HTTP GET request to the URL
try:
    response = requests.get(url)
    # Raise an HTTPError for bad responses (4xx or 5xx)
    response.raise_for_status()

    print(f"Successfully fetched {url} (Status Code: {response.status_code})")

    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # --- Web Scraping Examples ---

    # Example 1: Get the title of the page
    page_title = soup.title.text if soup.title else 'No Title Found'
    print(f"\nPage Title: {page_title}")

    # Example 2: Find all quotes and their authors on the page
    print("\n--- Top 10 Quotes and Authors ---")
    # Each quote is within a <div> with class 'quote'
    quote_divs = soup.find_all('div', class_='quote')

    if quote_divs:
        for i, quote_div in enumerate(quote_divs):
            if i < 10: # Limit to top 10 for brevity
                text = quote_div.find('span', class_='text').text if quote_div.find('span', class_='text') else 'N/A'
                author = quote_div.find('small', class_='author').text if quote_div.find('small', class_='author') else 'N/A'
                print(f"{i+1}. \"{text}\" - {author}")
            else:
                break
    else:
        print("No quotes found on the page.")

    # Example 3: Find all tags and list their occurrences
    print("\n--- Popular Tags ---")
    tags_div = soup.find('div', class_='tags-box') # The box containing popular tags
    if tags_div:
        tags = tags_div.find_all('a', class_='tag')
        tag_names = [tag.text for tag in tags]
        from collections import Counter
        tag_counts = Counter(tag_names)
        for tag, count in tag_counts.most_common(5): # Show top 5 most common tags
            print(f"Tag: {tag}, Count: {count}")
    else:
        print("No popular tags found.")

except requests.exceptions.RequestException as e:
    print(f"An error occurred during the request: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
